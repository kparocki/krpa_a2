```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read from runs.csv
runs = pd.read_csv("runs.csv")
runs = runs.drop(columns = ["Name", "Source Type", "Source Name", "User", "Status", "Run ID", "Duration", "Start Time"])
#cut out polynomial degree 9
runs = runs[runs["polynomial degree"] != 9]
runs.head()
```

```python
#create a separate df for each regressor
linear = runs[runs['regressor'] == '(\'regressor\', LinearRegression())']
random_forest = runs[runs['regressor'] == '(\'regressor\', RandomForestRegressor(random_state=30))']
svr = runs[runs['regressor'] == '(\'regressor\', SVR())']
```

```python
#plot mean_MAE, mean_MSE and mean_R2 for each regressor
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(linear['polynomial degree'], linear['mean_MAE'], label = 'Linear Regression')
plt.plot(random_forest['polynomial degree'], random_forest['mean_MAE'], label = 'Random Forest')
plt.plot(svr['polynomial degree'], svr['mean_MAE'], label = 'SVR')
plt.xlabel('Polynomial Degree')
plt.ylabel('Mean MAE')
plt.legend()

plt.subplot(1,3,2)
plt.plot(linear['polynomial degree'], linear['mean_MSE'], label = 'Linear Regression')
plt.plot(random_forest['polynomial degree'], random_forest['mean_MSE'], label = 'Random Forest')
plt.plot(svr['polynomial degree'], svr['mean_MSE'], label = 'SVR')
plt.xlabel('Polynomial Degree')
plt.ylabel('Mean MSE')
plt.legend()

plt.subplot(1,3,3)
plt.plot(linear['polynomial degree'], linear['mean_R2'], label = 'Linear Regression')
plt.plot(random_forest['polynomial degree'], random_forest['mean_R2'], label = 'Random Forest')
plt.plot(svr['polynomial degree'], svr['mean_R2'], label = 'SVR')
plt.xlabel('Polynomial Degree')
plt.ylabel('Mean R2')
plt.legend()
```
